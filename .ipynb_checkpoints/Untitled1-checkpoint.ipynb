# EDA

Na początek sprawdźmy ostatni, najważniejszy egzamin.

Uczniów podzielimy na cztery grupy zgodnie z procentowym podziałem maksymalnego wyniku 20 punktów. Słabi uczniowie osiągają wynik od 0 do 5 punktów, kiepscy od 6 do 10, dobrzy od 11 do 15, najlepsi od 16 do 20.

Na potrzeby EDA, stwórzmy nową cechę odwzorowującą poziom ucznia.

```python
def finished(x):
    if x >= 6 and x < 11:
        return 1
    if x >= 11 and x < 16:
        return 2
    if x >= 16:
        return 3
    else:
        return 0
```

```python
math = data.copy()
```

```python
math_y = math['G3']
```

```python
math['student_lvl'] = math['G3'].apply(finished)
```

Sprawdźmy jak rozkładają się wyniki egzaminu końcowego ostatniej klasy.

```python
plt.figure(figsize = (14,8))
ax = sns.histplot(data=math, x='G3', multiple='dodge', shrink=0.9, discrete=True)
ax.bar_label(ax.containers[0])
```

Najwięcej uczniów uzyskało wynik 10 punktów. Mamy też sporo niezdanych egzaminów. Większość danych znajduje się po prawej stronie wykresu.

```python
ax = sns.histplot(data=math, x='student_lvl', multiple='dodge', shrink=0.9, discrete=True)
```

Bazowy rozkład danych przedstawiający poziom uczniów.

Sprawdźmy jak poszczególne cechy wyglądają na tle ostatecznego poziomu ucznia.

## Relacja cechy z G3

```python
def plot_features(column=None):
    
    
    plt.figure(figsize = (14,8))
    ax = sns.histplot(data=math, x=column, hue='student_lvl', palette='Dark2', multiple='dodge', shrink=0.9, discrete=True)
    for container in ax.containers:
        ax.bar_label(container)

    
    x, y = column, 'student_lvl'
    df1 = math.groupby(x)[y].value_counts(normalize=True)
    df2 = df1.mul(100)
    df2 = df2.rename('percent').reset_index()
    
    #plt.figure(figsize = (14,8))
    g = sns.catplot(x=x,y='percent',hue=y,kind='bar',data=df2)
    plt.gcf().set_size_inches(18, 8)

    g.ax.set_ylim(0,105)

    for p in g.ax.patches:
        txt = str(p.get_height().round()) + '%'
        txt_x = p.get_x() + 0.01
        txt_y = p.get_height().round() + 0.01
        g.ax.text(txt_x,txt_y,txt)
        
    plt.show()
```

```python
# Wiek
plot_features(column='age')
```

W poszczególnych grupach wiekowych przeważają uczniowie uczący się dobrze. Wyjątkiem są uczniowie w wieku 18 lat u których widać przewagę uczących się kiepsko.

## Relacja cechy z przebiegiem nauki

Wiemy już jak wygląda nasz zbiór danych pod względem ostatecznego wyniku uczniów. Sprawdźmy teraz jak wygląda, jeśli za wartość przyjmiemy różnicę wyników pomiędzy pierwszym a ostatnim egzaminem. Uzyskana wartość przedstawia progres/regres ucznia. Dodatkowym parametrem będzie poziom ucznia.

Im większa wartość zmiany, tym większy progres zanotowali uczniowie z poszczególnej grupy.

Grupa odnotywująca słabe wyniki w nauce będzie przedstawiać mocny regress. Jest to spowodowane otrzymywanymi, zerowymi wynikami przez niektórych uczniów. Jeżeli uczeń zdobył 7 punktów na egzaminie G1 a z egzaminu G3 otrzymał 0, to jego regres będzie wynosił -7, co mocno wpływa na wygląd wykresu.

Dlatego pominę opisywanie tej właśnie grupy.

Ustalmy skalę progresu, żeby móc określić jego znaczenie. Znaczący progres ucznia można określiś w skali od 1 do 10. 2 punktowy progres wskazuje na skalę 1, natomiast 20 punktowy, na skalę 10.
Najwyższą możliwą wartością progresu jest 20. Nastąpiła by w momencie kiedy uczeń otrzymawszy 0 punktów na egzaminie G1 zdobyłby 20 na G3. Jednakże jest to niemożliwe.

```python
math_v2 = math.copy()
math_v2['difference'] = math_v2['G3'] - math_v2['G1']
```

```python
plt.figure(figsize=(11,7))
ax = sns.histplot(data=math_v2, x='difference')
ax.bar_label(ax.containers[0])
plt.show()
```

Najwięcej uczniów nie wykazało zmian w wynikach. Bardzo duża część zbioru wykazuje się regresem w nauce.

```python
plt.figure(figsize=(11,7))
ax = sns.barplot(data=math_v2, x='student_lvl', y='difference')
ax.bar_label(ax.containers[0])
plt.show()
```

Bazowy rozkład danych. Wraz z poziomem ucznia rośnie również jego progres w nauce.

```python
def a(column):
    plt.figure(figsize=(16,8))
    ax = sns.barplot(data=math_v2, x=column, y='difference', hue='student_lvl')
    for container in ax.containers:
        ax.bar_label(container, fmt='%1.2f', padding=10)
        
    plt.show()
    
    print(f'Liczebność grupy %: \n{round(math_v2[column].value_counts(normalize=True), 2)}')
```

# Klasyfikacja

```python
# Import bibliotek
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.pipeline import FeatureUnion, Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, PolynomialFeatures

from sklearn.metrics import classification_report, confusion_matrix, roc_curve, f1_score, roc_auc_score
```

```python
# Import modeli

from sklearn.linear_model import LogisticRegression

from sklearn.naive_bayes import CategoricalNB

from sklearn.svm import SVC
```

Klasyfikacji dokonamy przy pomocy trzech modeli, które nadają się do przetwarzania małych baz danych jak nasza. Z racji, że uczniowie zostali podzieleni na 4 grupy, użyjemy wariantów tych klasyfikatorów pozwalających nam na klasyfikację wieloklasową.

Logistic Regression

SVM

Categorical NB

Do oceny naszego modelu użyjemy metryk F1 Score oraz ROC AUC oraz Confussion Matrix.

## Przygotowanie danych

Na początek wrzucimy do modelu wszystkie cechy, czym uzyskamy wynik bazowy naszego modelu.

```python
df = math.copy()
```

Pozbądźmy się wartości tekstowych z naszego zbioru

```python
val_dict = {"school":    {"GP": 0, "MS": 1},
            "sex":       {"M": 0, "F": 1,},
            "address":   {"U": 0, "R": 1},
            "famsize":   {"LE3": 0, "GT3": 1},
            "Pstatus":   {"T": 0, "A": 1},
            "Mjob":      {"teacher": 0, "health": 1, "services": 2, "at_home": 3, "other": 4},
            "Fjob":      {"teacher": 0, "health": 1, "services": 2, "at_home": 3, "other": 4},
            "reason":    {"home": 0, "reputation": 1, "course": 2, "other": 3},
            "guardian":  {"mother": 0, "father": 1, "other": 2},
            "schoolsup": {"no": 0, "yes": 1},
            "famsup":    {"no": 0, "yes": 1},
            "paid":      {"no": 0, "yes": 1},
            "activities":{"no": 0, "yes": 1},
            "nursery":   {"no": 0, "yes": 1},
            "higher":    {"no": 0, "yes": 1},
            "internet":  {"no": 0, "yes": 1},
            "romantic":  {"no": 0, "yes": 1},
            }           
```

```python
df = df.replace(val_dict)
```

Danych jest naprawdę niewiele, dlatego pominiemy dane walidacyjne.

```python
X = df.drop(['student_lvl'], axis=1)
y = df['student_lvl']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Podzielmy zbiór na dane numeryczne i kategoryczne.

```python
cat_feat = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
           'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime', 'failures', 'schoolsup',
           'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic',
           'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health']

num_feat = ['age', 'absences', 'G1', 'G2', 'G3']
```

Stworzymy pipeline z preprocessingiem danych oraz treningiem modelu

```python
num_prep = Pipeline(steps=[
    ('polynomial_features', PolynomialFeatures(degree=3)),
    ('scaler_1', StandardScaler())
])

cat_prep = Pipeline(steps=[
    ('encoder', OneHotEncoder(sparse=False ,handle_unknown='ignore'))
])

data_prep = ColumnTransformer(transformers=[
    ('numeric_preprocessing', num_prep, num_feat),
    ('categorical_preprocessing', cat_prep, cat_feat)
])

log_reg = Pipeline(steps=[('preprocessor', data_prep),
                          ('model', LogisticRegression(multi_class='multinomial', solver='lbfgs'))])

svm = Pipeline(steps=[('preprocessor', data_prep),
                        ('model', SVC(probability=True))])
```

## Logistic Regression

```python
params = {
    'preprocessor__numeric_preprocessing__polynomial_features__degree': [1, 2, 3, 4],
    'model__C': [0.2, 0.5, 0.7, 1.0]
}

grid_search = GridSearchCV(log_reg, params, n_jobs=-1, verbose=10, scoring='f1_macro')
grid_search.fit(X_train, y_train)
print('Hiperparameters: ', grid_search.best_params_)
best_log = grid_search.best_estimator_
```

```python
preds = best_log.predict(X_test)
preds_proba = best_log.predict_proba(X_test)
```

```python
print(classification_report(y_test, preds))
```

```python
plt.figure()
cm = confusion_matrix(y_test, preds)
ax = sns.heatmap(cm, annot=True, cmap='Blues', fmt='.0f')
ax.set_title('Confusion Matrix\n\n')
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ')
plt.show()
```

```python
print(f1_score(y_test, preds, average='macro'))
print(roc_auc_score(y_test, preds_proba, multi_class='ovr'))
```

Model poradził sobie bardzo dobrze z klasyfikacją. Ma lekkie problemy z klasyfikacją uczniów kategorii 1, oraz 2 (kolejno, kiepscy oraz dobrzy).

## SVM

```python
params = {
    'preprocessor__numeric_preprocessing__polynomial_features__degree': [1, 2, 3, 4],
    'model__C': [0.2, 0.5, 0.7, 1.0],
    'model__kernel': ['linear', 'poly', 'rbf', 'sigmoid', 'precomputed']
}

grid_search = GridSearchCV(svm, params, n_jobs=-1, verbose=10, scoring='f1_macro')
grid_search.fit(X_train, y_train)
print('Hiperparameters: ', grid_search.best_params_)
best_svc = grid_search.best_estimator_
```

```python
preds_svm = best_svc.predict(X_test)
preds_proba_svm = best_svc.predict_proba(X_test)
```

```python
print(classification_report(y_test, preds_svm))
```

```python
plt.figure()
cm = confusion_matrix(y_test, preds_svm)
ax = sns.heatmap(cm, annot=True, cmap='Blues', fmt='.0f')
ax.set_title('Confusion Matrix\n\n')
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ')
plt.show()
```

```python
print(f1_score(y_test, preds_svm, average='macro'))
print(roc_auc_score(y_test, preds_proba_svm, multi_class='ovo'))
```

Model poradził sobie nieznacznie lepiej niż regresja logistyczna.

### Najlepszy model z użyciem tylko danych numerycznych.

Jak widać brak danych o wynikach egzaminów nie pozwala na prawidłową klasyfikację uczniów.

Czy model wykorzystujący TYLKO te dane poradzi sobie lepiej?

Sprawdźmy to na przy użyciu modelu SVC, który uzyskał najlepszy wynik.

```python
num_data = df.copy()
X_n = df[['G1', 'G2', 'G3']]
y_n = num_data['student_lvl']
X_train_n, X_test_n, y_train_n, y_test_n = train_test_split(X_n, y_n, test_size=0.3, random_state=42)
```

```python
num_feat_n = ['G1', 'G2', 'G3']

num_prep_n = Pipeline(steps=[
    ('polynomial_features', PolynomialFeatures(degree=3)),
    ('scaler_1', StandardScaler())
])

data_prep_n = ColumnTransformer(transformers=[
    ('numeric_preprocessing', num_prep_n, num_feat_n)
])

svm_v2 = Pipeline(steps=[('preprocessor', data_prep_n),
                        ('model', SVC(probability=True))])
```

```python
params_n = {
    'preprocessor__numeric_preprocessing__polynomial_features__degree': [1, 2, 3, 4],
    'model__C': [0.2, 0.5, 0.7, 1.0],
    'model__kernel': ['linear', 'poly', 'rbf', 'sigmoid', 'precomputed']
}

grid_search = GridSearchCV(svm_v2, params_n, n_jobs=-1, verbose=10, scoring='f1_macro')
grid_search.fit(X_train_n, y_train_n)
print('Hiperparameters: ', grid_search.best_params_)
new_svm = grid_search.best_estimator_
```

```python
preds_new_svm = new_svm.predict(X_test_n)
preds_proba_new_svm = new_svm.predict_proba(X_test_n)
```

```python
print(classification_report(y_test_n, preds_new_svm))
```

```python
plt.figure()
cm = confusion_matrix(y_test_n, preds_new_svm)
ax = sns.heatmap(cm, annot=True, cmap='Blues', fmt='.0f')
ax.set_title('Confusion Matrix\n\n')
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ')
plt.show()
```

```python
print(f1_score(y_test_n, preds_new_svm, average='macro'))
print(roc_auc_score(y_test_n, preds_proba_new_svm, multi_class='ovo'))
```

Oczywiście, że poradzi sobie lepiej! Jak wiemy, ostateczny poziom ucznia jest stworzony z wyników G3 a te, są mocno uzależnione od poprzednich egzaminów.
